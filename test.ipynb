```python
import TexSoup
print(f"TexSoup version: {TexSoup.__version__}")

# Let's try parsing just the problematic section
problematic_section = r"""\titleformat{\section}{
  \vspace{-4pt}\scshape\raggedright\large
}{}{0em}{}[\color{black}\titlerule \vspace{-5pt}]"""

print(f"\nTrying to parse problematic section:")
try:
    from TexSoup import TexSoup
    soup = TexSoup(problematic_section, tolerance=0)
    print("✓ Problematic section parses successfully")
except Exception as e:
    print(f"✗ Problematic section failed: {e}")

# Try with tolerance=1
try:
    soup = TexSoup(problematic_section, tolerance=1)
    print("✓ Problematic section parses with tolerance=1")
except Exception as e:
    print(f"✗ Problematic section failed even with tolerance=1: {e}")
```
